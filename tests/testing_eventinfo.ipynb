```python
def locate_event(data_dir,files,run,subrun,event):
  for i,file in enumerate(files):
    start = time()
    tree = uproot.open(f'{DATA_DIR}{file}')
    keys = tree.keys()
    rec_tree_name = tree.keys()[0]
    run_dict = tree[rec_tree_name].arrays(hdrkeys,library='np')
    run_list = np.array([run_dict[hdrkeys[0]],run_dict[hdrkeys[1]],run_dict[hdrkeys[2]]])
    for ele in run_list.T:
      if (collections.Counter([run,subrun,event]) == collections.Counter(list(ele))):
        print(f'In {file} : {time()-start:.2f} (s)')
        return files[i],list(run_list.T)
    #print(f'Not in {file} : {time()-start:.2f} (s)')
  return None
```
